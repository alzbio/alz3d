```python
import numpy as np
from pyrosetta import *
from pyrosetta.toolbox import cleanATOM
from pyrosetta import PyMOLMover
init()
fr=rosetta.protocols.relax.FastRelax()
scorefxn = get_score_function()
fr.set_scorefxn(scorefxn)
s24=pose_from_pdb("pdb/s24.pdb")
#fr.max_iter(100)
def mutate(mut,wt):
    loc=int(mut[1:-1])
    to_a=mut[-1:]
    loc=wt.pdb_info().pdb2pose('B', loc)
    m=wt.clone()
    pyrosetta.toolbox.mutants.mutate_residue(m,loc,to_a)
    fr.apply(m)
    m.dump_pdb('pdb/'+mut+'.pdb')
    return m
```

```python
%%time
wt=s24.clone()
fr.apply(wt)
wt.dump_pdb('pdb/wt.pdb')
wt.scores
```

```python
%%time
for mut in ['A99A','I143T','S169L','E280A','L166P','E318G','P117L','T147I','R278I','G384A','I439V','V94M']:
    print ('--------------------------------------------------------------------------------------------------------------->',mut)
    m=mutate(mut,wt)
```
